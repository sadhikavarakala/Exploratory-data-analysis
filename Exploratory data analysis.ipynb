```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset

file_path = '/Users/siddharthsunkam/Desktop/Sleep_health_and_lifestyle_dataset.csv'
data = pd.read_csv(file_path)
# Display basic information and the first few rows of the dataset
print(data.info())
print(data.head())

# Generate descriptive statistics for the dataset
print(data.describe())
print(data.describe(include=['object']))

# Visualizations
# Setting the aesthetic style of the plots
sns.set(style="whitegrid")

# Distribution of numerical variables
fig, ax = plt.subplots(3, 3, figsize=(15, 15))
numerical_columns = ['Age', 'Sleep Duration', 'Quality of Sleep', 'Physical Activity Level', 'Stress Level', 'Heart Rate', 'Daily Steps']
for i, col in enumerate(numerical_columns):
    sns.histplot(data[col], kde=True, ax=ax[i//3, i%3]).set_title(f'Distribution of {col}')

# Distribution of categorical variables
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
categorical_columns = ['Gender', 'BMI Category', 'Sleep Disorder']
for i, col in enumerate(categorical_columns):
    sns.countplot(x=col, data=data, ax=ax[i]).set_title(f'Distribution of {col}')

plt.tight_layout()

# Boxplots to examine the distribution of sleep duration across different categories
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
sns.boxplot(x='BMI Category', y='Sleep Duration', data=data, ax=ax[0]).set_title('Sleep Duration by BMI Category')
sns.boxplot(x='Gender', y='Sleep Duration', data=data, ax=ax[1]).set_title('Sleep Duration by Gender')

# Correlation matrix
corr_matrix = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix')

# Scatter plot to examine the relationship between two variables
plt.figure(figsize=(7, 5))
sns.scatterplot(x='Stress Level', y='Quality of Sleep', data=data).set_title('Stress Level vs. Quality of Sleep')

plt.tight_layout()
plt.show()
```
